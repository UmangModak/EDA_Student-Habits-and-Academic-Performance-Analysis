# EDA : Exploratory data analysis : EDA, or Exploratory Data Analysis, is a crucial first step in data analysis where you use visualization and statistical methods to understand a dataset's main characteristics, detect anomalies, and identify relationships between variables.

 1.Univariant Analysis :it is use to plot or visuals  withn one columns.
 2.Multivarient Analysis  / Bivariant : it has mmultiple input

## UNivariant analysis

```python
a = sns.countplot(x='gender',data=df)
plt.figure(figsize=(12,6))
for bars in a.containers:
    a.bar_label(bars)
```

from the above graph we can see that the total number of male ,female and other are follows as 440 , 433 and 36 . we can see that the male count / percentage is slightly higher than female and other.

```python
a= sns.countplot(x='age',data= df,hue='gender')
plt.figure(figsize=(12,10))

for bars in a.containers:
    a.bar_label(bars)
```

from the above graph it is clearly shown that the 20 age group has the highest participant  containing male = 60 , female 70 and other = 7 
the female participant is slighlyt higher in mostly age group.
